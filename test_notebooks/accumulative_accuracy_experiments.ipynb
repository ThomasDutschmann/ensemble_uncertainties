### Validate Calculation of ROC and Cumulative Accuracy

Import packages

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from sklearn.feature_selection import VarianceThreshold
from sklearn.metrics import accuracy_score, roc_curve
from sklearn.model_selection import KFold
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
```

Settings

```python
mpl.rcParams['figure.dpi'] = 200
red = [1, 0, 0]
n_splits = 10
data_dir = '../test_data/CYP1A2_klingspohn/'
subset = 'CYP1A2'
```

Load datasets

```python
X = pd.read_csv(f'{data_dir}{subset}_autoscaled_X.dat', sep='\s+', header=None, dtype=np.float32)
y = pd.read_csv(f'{data_dir}{subset}_Y.dat', sep='\s+', header=None, dtype=np.int32)
X['id'] = [str(i) for i in range(len(X))]
y['id'] = X['id']
y.columns = ['y', 'id']
X = X.set_index('id').sort_index()
y = y.set_index('id').sort_index()
```

Prepare prediction collectors

```python
test_predictions_list = list()
test_probabilities_list = list()
```

Acquire predictions, collect probabilities

```python
kfold = KFold(n_splits=n_splits, random_state=0, shuffle=True)
splits = kfold.split(X) 

for split_index, (train_id, test_id) in enumerate(splits):
    print(f'At split {split_index+1}.', end=' ')
    # Select train and test fraction
    X_tr, X_te = X.iloc[train_id], X.iloc[test_id]
    y_tr, y_te = y.iloc[train_id], y.iloc[test_id]
    # Set scaler and transformer
    vt = VarianceThreshold().fit(X_tr)
    scaler = StandardScaler().fit(vt.transform(X_tr))
    # Fit and predict
    X_train = pd.DataFrame(scaler.transform(vt.transform(X_tr)),
        index=X_tr.index, columns=X_tr.columns[vt.get_support()])
    # Variance-filter and scale test inputs
    X_test = pd.DataFrame(scaler.transform(vt.transform(X_te)),
        index=X_te.index, columns=X_te.columns[vt.get_support()])
    # Construct model (fit)
    model = SVC(probability=True)
    model.fit(X_train, y_tr.values.ravel())
    # Predict
    predictions = model.predict(X_test)
    print(f'Accuracy: {accuracy_score(y_te["y"], predictions):.3f}.')
    te_preds = pd.DataFrame(predictions, index=X_test.index)
    te_probs = pd.DataFrame(model.predict_proba(X_test), index=X_test.index)
    test_predictions_list.append(te_preds)
    test_probabilities_list.append(te_probs)

# Concatenate single frames
test_predictions = pd.concat(test_predictions_list).sort_index()
test_probabilities = pd.concat(test_probabilities_list).sort_index()
```

Evaluate

```python
print(f'Accuracy: {accuracy_score(y, test_predictions):.3f}')
```

Plot AUC

```python
def plot_roc(trues, probs):
    fpr, tpr, _ = roc_curve(trues, probs)
    # Plot
    plt.figure(figsize=(5, 5))
    plt.grid(zorder=1000)
    plt.plot(fpr, tpr, color=red, zorder=100, label='ROC')
    plt.plot([0.0, 1.0], [0.0, 1.0], zorder=100,
        color='k', label='Random', linestyle='dashed')    
    plt.ylabel('True positive rate')
    plt.xlabel('False positive rate')
    plt.legend()
    plt.plot()
    

plot_roc(y, test_probabilities[1])
```

Compute and plot cumulative accuracy

```python
def cumulative_accuracy(trues, predicted, probabilities, start_frac=.05):
    # Sort by probabilities
    certainties = pd.Series(probabilities.values - .5,
        index=probabilities.index).abs()
    sorted_certainties = certainties.sort_values(ascending=False)
    sorted_pred = list(predicted.reindex(sorted_certainties.index).values)
    # Sort true outputs
    sorted_y = list(trues.reindex(sorted_certainties.index).values)
    # Initialize
    accuracies = list()
    # Compute
    for i in range(int(start_frac*len(trues)), len(trues+1)):
        y_partition = sorted_y[:i]
        pred_partition = sorted_pred[:i]
        current_accuracy = accuracy_score(y_partition, pred_partition)
        accuracies.append(current_accuracy)
    return accuracies


def plot_cumulative_accuracy(trues, predicted, probabilities, start_frac=.05):
    accuracies = cumulative_accuracy(trues, predicted, probabilities, start_frac=start_frac)
    x_space = np.linspace(100*start_frac, 100.0, len(accuracies))
    # Plot
    plt.figure(figsize=(5, 5))
    plt.grid(zorder=1000)
    plt.plot(x_space, accuracies, color=red, zorder=100)
    plt.xlabel('Percentile')
    plt.ylabel('Cumulative accuracy')
    plt.show()
                   

plot_cumulative_accuracy(y, test_predictions, test_probabilities[1])
```
